```python
import pandas as pd
import numpy as np
```

## WTI

```python
df = pd.read_excel("Stationary Data/WTI.xlsx") 
df
```

```python
df.rename(columns = {'MCOILWTICO':'WTI'}, inplace = True) 
df.rename(columns = {'MCOILWTICO_LAG':'WTI_LAG'}, inplace = True) 

df.rename(columns = {'EMVOVERALLEMV':'Equity Market Volatility'}, inplace = True) 
df.rename(columns = {'EUEPUINDXM':'Economic Policy Uncertainty Europe'}, inplace = True) 
df.rename(columns = {'GACDISA066MSFRBNY':'NY Business Conditions'}, inplace = True)
df.rename(columns = {'TB3MS':'3 Month Treasury Bill'}, inplace = True) 
df.rename(columns = {'INFECTDISEMVTRACK':'Infectious Disease Tracker'}, inplace = True) 
df.rename(columns = {'PNICKUSDM':'Nickel'}, inplace = True) 

df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)
df
```

```python
df.to_excel("Renamed Columns Stationary Data/WTI.xlsx", index=True)
print("DataFrame saved")
```
