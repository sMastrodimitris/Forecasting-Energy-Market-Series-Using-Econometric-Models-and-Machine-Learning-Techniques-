```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid", color_codes=True)
from scipy.stats import skew, kurtosis

```

## WTI

```python
def analyze_column(path,column_name):
    
    df = pd.read_excel(path) 
    df['DATE'] = pd.to_datetime(df['DATE'])
    df.set_index('DATE', inplace=True)
    df 
    
    # Basic Statistics
    sample_size = len(df[column_name])
    std_dev = df[column_name].std()
    skewness = skew(df[column_name])
    kurt = kurtosis(df[column_name])

    print(f"Sample Size: {sample_size}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"Skewness: {skewness:.2f}")
    print(f"Kurtosis: {kurt:.2f}")

    # Histogram
    plt.figure(figsize=(12, 6))
    plt.hist(df[column_name], bins=30, edgecolor='k', alpha=0.7)
    plt.title('Histogram')
    plt.show()

    # Boxplot
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=df[column_name], color='green')
    plt.title('Boxplot')
    plt.show()
```

```python
analyze_column("WTI/Monthly/Renamed Columns Stationary Data/WTI.xlsx","WTI")
```

```python
analyze_column("WTI/Quarterly/Renamed Columns Stationary Data/WTI.xlsx","WTI")
```
