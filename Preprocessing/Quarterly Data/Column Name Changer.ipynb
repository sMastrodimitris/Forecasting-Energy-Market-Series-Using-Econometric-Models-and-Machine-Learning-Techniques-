```python
import pandas as pd
import numpy as np
```

## WTI

```python
df = pd.read_excel("Stationary Data/WTI.xlsx") 
df
```

```python
df.rename(columns = {'POILWTIUSDQ':'WTI'}, inplace = True) 
df.rename(columns = {'POILWTIUSDQ_LAG1':'WTI_LAG1'}, inplace = True) 
df.rename(columns = {'POILWTIUSDQ_LAG2':'WTI_LAG2'}, inplace = True) 
df.rename(columns = {'POILWTIUSDQ_LAG3':'WTI_LAG3'}, inplace = True) 
df.rename(columns = {'POILWTIUSDQ_LAG4':'WTI_LAG4'}, inplace = True) 

df.rename(columns = {'GDP':'USAGDP'}, inplace = True) 
df.rename(columns = {'EMVOVERALLEMV':'Equity Market Volatility'}, inplace = True) 
df.rename(columns = {'TB3MS':'3 Month Treasury Bill'}, inplace = True) 
df.rename(columns = {'INFECTDISEMVTRACK':'Infectious Disease Tracker'}, inplace = True) 

df['DATE'] = pd.to_datetime(df['DATE'])
df.set_index('DATE', inplace=True)
df
```

```python
df.to_excel("Renamed Columns Stationary Data/WTI.xlsx", index=True)
print("DataFrame saved")
```
